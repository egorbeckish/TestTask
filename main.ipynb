```python
import pandas as pd
import numpy as np
```

```python
financial_data = pd.read_csv("financial_data.csv")
prolongations = pd.read_csv("prolongations.csv")
```

```python
prolongations
```

```python
# Получение название колонок, которые являются месяцами

months_col = [
	col 
	for col in financial_data.columns 
	if col not in ['id', 'Причина дубля', 'Account']
]
months_col
```

```python
# Группируем ответственных с месяцами

financial_data_melted = financial_data.melt(
    id_vars=["id", "Account"],
    value_vars=months_col,
    var_name="month",
)
financial_data_melted
```

```python
# функция обработки данных

def clear_value(value):
	if isinstance(value, str):
		if value == "в ноль":
			return 0
		
		if value in ["end", "стоп"]:
			return np.nan
		
		if "," in value:
			return float("".join(value.replace(",", ".").split()))
	
	if isinstance(value, float):
		return value
```

```python
# Преобразование данных в нужный вид, с учетом поставленных условий

financial_data_melted["clear_value"] = financial_data_melted["value"].apply(clear_value).copy()
financial_data_melted
```

```python
# Удаление неизвестных данных

financial_data_melted = financial_data_melted.dropna()
financial_data_melted
```

```python
# Соединяем таблицы с очищенными данными через левое соединение, чтобы учитывать все проекты

df_pr_fi = pd.merge(
	prolongations, 
	financial_data_melted, 
	how="left",
	left_on=["id", "AM"], 
	right_on=["id", "Account"]
)
```

```python
# Функция для приведения месяцев формата str -> datetime

def replace_month(month):
	month, year = month.split()
	month = {
		"январь": "01",
		"февраль": "02",
		"март": "03",
		"апрель": "04",
		"май": "05",
		"июнь": "06",
		"июль": "07",
		"август": "08",
		"сентябрь": "09",
		"октябрь": "10",
		"ноябрь": "11",
		"декабрь": "12",
	}[month]
	
	return f"{year}-{month}"
```

```python
df_pr_fi["month"] = df_pr_fi["month_x"].apply(replace_month)
df_pr_fi.drop(columns=["month_x", "month_y"], inplace=True)
df_pr_fi["month"] = pd.to_datetime(df_pr_fi["month"], format="%Y-%m")
df_pr_fi
```

```python
# Берем только уникальные месяцы, чтобы не повторятсья

months = df_pr_fi["month"].unique()
months
```

```python
# Функция для вычисления предыдущего месяца и 2 месяца назад

def prev_date(date):
	day, month, year = date.day, date.month, date.year

	if month - 1 == 0:
		return pd.Timestamp(f"{year - 1}-12-{day}"), pd.Timestamp(f"{year - 1}-11-{day}")
	
	if month - 2 == 0:
		return pd.Timestamp(f"{year}-{month - 1}-{day}"), pd.Timestamp(f"{year - 1}-12-{day}")
	
	return pd.Timestamp(f"{year}-{month - 1}-{day}"), pd.Timestamp(f"{year}-{month - 2}-{day}")
```

```python
results = []
for month in months:
	# Находим ответственных по предыдущему месяцу
	managers = df_pr_fi[df_pr_fi["month"] == month]["AM"].unique()
	
	for manager in managers:
		# даты предыдущих месяцев у ответственного
		prev, prev2 = prev_date(month)
		
		# Проекты в предыдущем месяце
		prev_project = df_pr_fi[
			(df_pr_fi["AM"] == manager) &
			(df_pr_fi["month"] == prev)
		]["id"].unique()
		
		if prev_project.shape[0] == 0:
			continue
		
		# Сумма за проекты в предыдущем месяце
		prev_total = df_pr_fi[
			(df_pr_fi["id"].isin(prev_project)) &
			(df_pr_fi["month"] == prev)
		]["clear_value"].sum()

		# Сумма за проекты в текущем месяце
		curr_total = df_pr_fi[
			(df_pr_fi["id"].isin(prev_project)) &
			(df_pr_fi["month"] == month)
		]["clear_value"].sum()

		if prev_total == 0:
			coef1 = np.nan
		else:
			coef1 = curr_total / prev_total
		
		# Проекты 2 месяца назад
		prev2_project = df_pr_fi[
			(df_pr_fi["AM"] == manager) &
			(df_pr_fi["month"] == prev2)
		]["id"].unique()
		
		if prev2_project.shape[0] == 0:
			continue
		
		# Сумма за проекты 2 месяца назад
		prev2_total = df_pr_fi[
			(df_pr_fi["id"].isin(prev_project)) &
			(df_pr_fi["month"] == prev2)
		]["clear_value"].sum()

		# Сумма за проекты в текущем месяце
		curr2_total = df_pr_fi[
			(df_pr_fi["id"].isin(prev2_project)) &
			(df_pr_fi["month"] == month)
		]["clear_value"].sum()

		if prev2_total == 0:
			coef2 = np.nan
		else:
			coef2 = curr2_total / prev2_total

		results += [
			{
				"Месяц": month,
				"Ответственный": manager,
				"Коэфф. пролонгации за 1й месяц": coef1,
				"Коэфф. пролонгации за 2й месяц": coef2,
				"Cумма за последний месяц": prev_total,
				"Cумма за текущий месяц месяц": curr_total,
				"Cумма за последние 2 месяца": prev2_total,
				"Cумма за текущий месяц с учетом 2х месяцев": curr2_total,
			}
		]
```

```python
df_res = pd.DataFrame(results)
```

```python
# Функция расчет для среднего для года

def year_middle(data, col, total):
	if data[total].sum() == 0:
		return np.nan
	
	return np.average(data[col], weights=data[total])
```

```python
year_results = []
for manage in df_res["Ответственный"].unique():
	data = df_res[df_res["Ответственный"] == manage]

	coef1 = year_middle(data, "Коэфф. пролонгации за 1й месяц", "Cумма за последний месяц")
	coef2 = year_middle(data, "Коэфф. пролонгации за 2й месяц", "Cумма за последние 2 месяца")

	year_results += [
		{
			"Ответственный": manage,
			"Коеффициент 1": coef1,
			"Коеффициент 2": coef2,
		}
	]
```

```python
df_year_res = pd.DataFrame(year_results)
```

```python
# Расчет годовых данных

prev_total = df_res["Cумма за последний месяц"].sum()
curr_total = df_res["Cумма за текущий месяц месяц"].sum()

if prev_total == 0:
	coef1 = np.nan
else:
	coef1 = curr_total / prev_total


prev2_total = df_res["Cумма за последние 2 месяца"].sum()
curr2_total = df_res["Cумма за текущий месяц с учетом 2х месяцев"].sum()

if prev2_total == 0:
	coef2 = np.nan
else:
	coef2 = curr2_total / prev2_total
```

```python
df_year_res = pd.concat(
	[
		df_year_res,
		pd.DataFrame(
			{
				"Ответственный": ["Всего"],
				"Годовой коэфф. 1": [coef1],
				"Годовой коэфф. 2": [coef2],
			}
		)
	],
	ignore_index=True
)
```

```python

with pd.ExcelWriter("Отчет.xlsx") as file:
	df_res.to_excel(file, sheet_name="Месячный", index=False)
	df_year_res.to_excel(file, sheet_name="Годовой", index=False)
```
